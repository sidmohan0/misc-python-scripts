```python
# Initialize an empty dictionary to store the missing value counts for each column
missing_values_count = {}

# Iterate over all columns in the dataframe
for column in df.columns:
    # Count the number of missing values in the current column
    missing_values = df[column].isnull().sum()
    # Add the missing value count to the dictionary
    missing_values_count[column] = missing_values
    
# Check if there are any missing values in the dataframe
if any(missing_values_count.values()):
    print("There are missing values in the dataframe.")
    print(missing_values_count)
    # Get the rows that contain missing values
    missing_rows = df[df.isnull().any(axis=1)]
    print(missing_rows)
else:
    print("There are no missing values in the dataframe.")
```
